```python
# expected value of inverse gamma precision also needs expected value of log and original
def black_box_variational_inference(logprob, num_biases):

    def unpack_params(params):
        
        mean, log_std, tau, log_tau_std, nu, log_nu_std, mean_bias, log_std_bias, preci_mu, preci_log_std,c_mu,c_log_std= params[:n1],params[n1:n2],\
        params[n2:n3], params[n3:n4], params[n4:n5],params[n5:n6], params[n6:(n6+num_biases)],params[(n6+num_biases):-4],params[-4:-3],params[-3:-2],params[-2:-1],params[-1:]
         
        lambda_a_q, theta_a_q, lambda_b_q, theta_b_q = fixed_update(tau, weird(log_tau_std), nu, weird(log_nu_std))
        
        return mean, log_std, tau, log_tau_std, nu, log_nu_std, mean_bias, log_std_bias, preci_mu, preci_log_std,lambda_a_q, theta_a_q, lambda_b_q, theta_b_q, c_mu, c_log_std
    
    def variational_objective(params, t):
                        
        mean, log_std, tau, log_tau_std, nu, log_nu_std, mean_bias, log_std_bias, preci_mu, preci_log_std,lambda_a_q, theta_a_q, lambda_b_q, theta_b_q, c_mu, c_log_std = unpack_params(params)
        
        preci_std, c_std = weird(preci_log_std), weird(c_log_std)
        
        Elog_sigma = preci_mu
        E_sigma = np.exp(0.5 * preci_std ** 2 + preci_mu)
        
        sigma_tau, sigma_nu, sigma_mean, sigma_bias= weird(log_tau_std), weird(log_nu_std), weird(log_std), weird(log_std_bias)
        
        lambda_a_q, theta_a_q, lambda_b_q, theta_b_q = fixed_update(tau, sigma_tau, nu, sigma_nu)
        
        lower_bound = wholesale_entropy(log_std, tau, log_tau_std, nu, log_nu_std, lambda_a_q,\
                      lambda_b_q, theta_a_q, theta_b_q, n1) + Ex_Gaussian(mean[:-num_output_w],sigma_mean[:-num_output_w],tau2) + \
                      logprob(mean, mean_bias, sigma_mean,sigma_bias,tau,nu,sigma_tau,sigma_nu, Elog_sigma, E_sigma,c_mu,c_std, t) + \
                      Ecauchy(tau_a, tau, sigma_tau, lambda_a, b_0, lambda_a_q, lambda_b_q) + \
                      Ecauchy(nu_a, nu, sigma_nu, theta_a, b_g, theta_a_q, theta_b_q) + \
                      gaussian_entropy(log_std_bias, num_biases) +\
                      Ex_Gaussian(mean_bias,sigma_bias,tau2_bias) + \
                      Ex_Gaussian(mean[-num_output_w:], sigma_mean[-num_output_w:],tau2_output) + \
                      Ex_lognormal_g(sigma_a, sigma_b, E_sigma, preci_mu) + log_normal_entropy(preci_log_std,preci_mu) +\
                      Ex_lognormal_ig(c_a, c_b, np.exp(0.5 * c_std ** 2 - c_mu),c_mu) + log_normal_entropy(c_log_std,c_mu)

        return -lower_bound
    
    gradient = grad(variational_objective)

    return variational_objective, gradient, unpack_params

def make_nn_funs(layer_sizes):

    shapes = list(zip(layer_sizes[:-1], layer_sizes[1:]))
    num_weights = sum(m*n for m, n in shapes)

    def unpack_layers(weights,bias_weights,weights_std,bias_std,tau,nu,tau_std,nu_std,c_mu,c_std):
        for layer,(m, n) in enumerate(shapes):
            if layer < n_l:
                yield weights[:m*n].reshape((m, n)),\
                      weights_std[:m*n].reshape((m, n)),\
                      bias_weights[:n],\
                      bias_std[:n],\
                      tau[:n],\
                      nu[layer],\
                      tau_std[:n],\
                      nu_std[layer],\
                      c_mu,\
                      c_std
                weights = weights[m*n:]
                weights_std = weights_std[m*n:]
                bias_weights = bias_weights[n:]
                bias_std = bias_std[n:]
                tau = tau[n:]
                tau_std = tau_std[n:]
            else:
                 yield weights[:m*n].reshape((m, n)),\
                      weights_std[:m*n].reshape((m, n)),\
                      bias_weights[:n],\
                      bias_std[:n],\
                      None,\
                      None,\
                      None,\
                      None,\
                      None,\
                      None
    #rs = npr.RandomState(5)         
    def predictions(weights,bias_weights,weights_std,bias_std,tau,nu,tau_std,nu_std,c_mu,c_std, inputs): 
        for layer, (W ,W_std, b, b_std,tau,nu,tau_std,nu_std,c_mu,c_std) in enumerate(unpack_layers(weights,bias_weights,weights_std,bias_std,tau,nu,tau_std,nu_std,c_mu,c_std)):
            if layer < n_l:
                tau_nu = 0.5 * (tau + nu + c_mu)
                tau_nu_var = 0.25 * (tau_std ** 2 + nu_std **2 + c_std **2)
                sample_tau_nu = np.exp(npr.randn(tau_nu.shape[0]) * np.sqrt(tau_nu_var) + tau_nu)
                mu = np.dot(inputs,W) + b
                var = np.dot(inputs **2, W_std **2) + b_std **2
                #outputs = (np.sqrt(var)) * rs.normal(size=mu.shape) + mu
                outputs = (np.sqrt(var) / np.sqrt(inputs.shape[1])) * npr.normal(size=mu.shape) + mu
                outputs = sample_tau_nu * outputs
                inputs = outputs * (outputs > 0)
            else:
                mu = np.dot(inputs,W) + b
                var = np.dot(inputs **2, W_std **2) + b_std **2
                #outputs = (np.sqrt(var)) * rs.normal(size=mu.shape) + mu
                outputs = (np.sqrt(var)/np.sqrt(inputs.shape[1])) * npr.normal(size=mu.shape) + mu
        return outputs
    
    
    def predictions2(weights,bias_weights,weights_std,bias_std,tau,nu,tau_std,nu_std,c_mu,c_std, inputs): 
        for layer, (W ,W_std, b, b_std,tau,nu,tau_std,nu_std,c_mu,c_std) in enumerate(unpack_layers(weights,bias_weights,weights_std,bias_std,tau,nu,tau_std,nu_std,c_mu,c_std)):
            if layer < n_l:
                tau_nu = 0.5 * (tau + nu + c_mu)
                tau_nu_var = 0.25 * (tau_std ** 2 + nu_std **2 + c_std **2)
                #sample_tau_nu = np.exp(tau_nu + 0.5 * tau_nu_var)   #mean
                sample_tau_nu = np.exp(tau_nu - 0.5 * tau_nu_var)    #mode
                mu = np.dot(inputs,W) + b
                outputs = mu
                outputs = sample_tau_nu * outputs
                inputs = outputs * (outputs > 0)
            else:
                outputs = np.dot(inputs,W) + b
        return outputs
    
    def logprob(weights,bias_weights,weights_std,bias_std,tau,nu,tau_std,nu_std, Elog_sigma, E_sigma,c_mu,c_std, inputs,targets):
        
        preds = predictions(weights,bias_weights,weights_std,bias_std,tau,nu,tau_std,nu_std,c_mu,c_std,inputs)
        log_lik = np.sum(-0.5 * np.log(pi2) + 0.5 * Elog_sigma - (0.5 * (targets - preds)**2 * E_sigma) )
        return log_lik 

    return shapes, num_weights, predictions, logprob,predictions2

def weird(x):
    #return np.log1p(np.exp(x))
    return np.exp(x)

def Ex_Gaussian(b_mean,b_sigma,tau_mf): 
    
    return np.sum(-0.5 * np.log(pi2) - 0.5 * np.log(tau_mf) - 0.5 / tau_mf * (b_mean ** 2 + b_sigma ** 2))
    
def Ecauchy(tau_a, tau, tau_std, a, b, a_q, b_q):
    
    l_ab = a_q / b_q
    E_log = np.log(b_q) - psi (a_q)
    
    Etau_lambda = - tau_a * E_log - gammaln(a) -\
                  (tau_a + 1.) * tau - np.exp( - tau + 0.5 * tau_std ** 2) * l_ab
      
    Elambda = -2 * a * np.log(b) - gammaln(a) -\
              (a + 1.) * E_log - (1./(b ** 2)) * l_ab
    
    return np.sum(Etau_lambda) + np.sum(Elambda)

def Ex_lognormal_g (prior_a, prior_b, E_tau, mu):
    return prior_a * np.log(prior_b) - gammaln(prior_a) + (prior_a - 1) * mu - prior_b * E_tau

def Ex_lognormal_ig (prior_a, prior_b, E_tau, mu):
    return prior_a * np.log(prior_b) - gammaln(prior_a) - (prior_a + 1) * mu - prior_b * E_tau

def fixed_update(tau, sigma_tau, nu, sigma_nu):
    
    lambda_a_q = theta_a_q = 1.
    lambda_b_q = 1.0 / (b_0 ** 2) + np.exp (- tau + 0.5 * sigma_tau ** 2)
    theta_b_q =  1.0 / (b_g ** 2) + np.exp (- nu + 0.5 * sigma_nu ** 2)
    
    return lambda_a_q, theta_a_q, lambda_b_q, theta_b_q

def wholesale_entropy(log_std, tau, log_tau_std, nu, log_nu_std, lambda_a_q,
                      lambda_b_q, theta_a_q, theta_b_q,num_weights):
    
    return gaussian_entropy(log_std, num_weights) + log_normal_entropy(log_tau_std, tau) + \
            log_normal_entropy(log_nu_std, nu) +  \
            inv_gamma_entropy(lambda_a_q, np.log(lambda_b_q)) + \
            inv_gamma_entropy(theta_a_q, np.log(theta_b_q)) 

def unpack_dummy(dummy):
    lambda_a_q, theta_a_q, lambda_b_q, theta_b_q = dummy[:1],dummy[1:2], dummy[:-1],dummy[-1:]
    return lambda_a_q, theta_a_q, lambda_b_q, theta_b_q
    
def inialization(shapes):
    
    rs = npr.RandomState(0)
    wlist = []
    for m, n in shapes:
        wlist.append(rs.randn(m * n) * np.sqrt(2 / m))
    w = np.concatenate(wlist)
    init_tau = np.log(np.abs(b_0 * rs.randn(num_tau_weights)/rs.randn(num_tau_weights)))
    init_nu = np.log(np.abs(b_g * rs.randn(n_l)/rs.randn(n_l)))
    bias_w = np.zeros(num_biases)
    
    scale = 5
    init_tau_log_std = -scale * np.ones(num_tau_weights) 
    init_log_std =  - scale * np.ones(n1)
    init_log_bias_std = - scale * np.ones(num_biases) 
    init_nu_std = - scale * np.ones(n_l)
    init_preci_mu = np.log(sigma_a/sigma_b)            
    init_preci_log_std = -scale 
    init_shrink_mu = np.log(c_b/(c_a + 1))
    init_shrink_log_std = -scale
    
    lambda_a_q, theta_a_q, lambda_b_q, theta_b_q = fixed_update(init_tau,weird(init_tau_log_std),init_nu,weird(init_nu_std))
    
    init_var_params = np.concatenate([w,init_log_std,init_tau,init_tau_log_std,init_nu,init_nu_std,bias_w,init_log_bias_std,[init_preci_mu,init_preci_log_std,init_shrink_mu,init_shrink_log_std]])
    return init_var_params,lambda_a_q, theta_a_q, lambda_b_q, theta_b_q

if __name__ == '__main__':
    #step0.005, regbias 0.001, regout0.01, sigma_a =1, sigma_b=1, b_0=1, b_0=1e-5##
    rbf = lambda x: np.exp(-x**2)
    toy = False
    init = False
    scale = True
    step = 0.005
    reg_bias = 0.001
    reg_output = 0.01
    data_seed = 5 ####0 difficult, 5 easy###
    
    tau2_bias = 0.5 / reg_bias
    tau2 = 1 ##non-center beta variance
    tau2_output = 0.5 / reg_output
    max_iter = 100000
    
    split_size = 0.1
    output_size= 1
    pi2 = 2. * np.pi
    lambda_a = theta_a = tau_a = nu_a = 0.5
    
    b_0 = 1
    b_g = 1e-4
    
    sigma_a = 1
    sigma_b = 0.01
    
    c_a = 1
    c_b = 0.01
    
    c_mu_master = c_std_master = preci_master =  preci_log_std_master = nu_master = nu_std_master = theta_master = lambda_master = mean_master = tau_std_master = tau_master = rmse_master = []
    
    if toy == True:
        x_len = 200
        noise = 0.1
        inputs, inputs_test, targets,  targets_test, dim, train_std, train_mu = build_toy_dataset(x_len,noise,split_size, seed=1990, state=data_seed)
        index_train = np.squeeze(np.argsort(inputs,axis=0))  
        index_test = np.squeeze(np.argsort(inputs_test,axis=0))   
    else:
        inputs, inputs_test, targets,  targets_test, dim, train_std, train_mu = build_boston_dataset(split_size,scale, state=data_seed)

    sup = [dim, 100, output_size]
    num_output_w = np.prod(sup[-2:])
    num_biases = np.sum(sup[1:])
    shapes, num_weights, predictions, logprob, predictions2 = \
    make_nn_funs(layer_sizes = sup)
    
    n1 = num_weights
    num_tau_weights =  np.sum(sup[1:(len(sup) - 1)])
    n2 = 2 * n1
    n3 = n2 + num_tau_weights
    n4 = n3 + num_tau_weights
    n_l = len(sup) - 1 - 1
    n5 = n4 + n_l
    n6 = n5 + n_l
    
    log_posterior = lambda weights,bias_weights,weights_std,bias_std,tau,nu,tau_std,nu_std, Elog_sigma, E_sigma,c_mu,c_std,t: logprob(weights,bias_weights,weights_std,bias_std,tau,nu,tau_std,nu_std,Elog_sigma, E_sigma,c_mu,c_std, inputs, targets)
    
    objective, gradient, unpack_params = \
        black_box_variational_inference(log_posterior,num_biases)
    print("Data size is",inputs.shape)
    print("Target size is",targets.shape)
    print('Bias regularizer is', reg_bias)
    print('Output regularizer is', reg_output)
    print('global shrinkage paramter is', b_g)
    print('regularzier paramter is',c_a,c_b) 
    def callback(params, t, g):
        
        if t % 500 == 0 or t == max_iter:
            mean, log_std, tau, log_tau_std, nu, log_nu_std, mean_bias, log_std_bias, preci_mu, preci_log_std,lambda_a_q, theta_a_q, lambda_b_q, theta_b_q,c_mu,c_log_std = unpack_params(params)
            sigma_tau, sigma_nu,c_std = weird(log_tau_std), weird(log_nu_std), weird(c_log_std)
            global log_a_master, log_b_master, rmse_master, tau_master, tau_std_master,mean_master,nu_master,nu_std_master,theta_master, lambda_master, preci_master, preci_log_std_master, c_mu_master,c_std_master
            outputs = predictions2(mean, mean_bias, np.zeros(num_weights), np.zeros(num_biases),\
                                  tau,nu,sigma_tau,sigma_nu,c_mu,c_std,inputs_test)
#             rs = npr.RandomState(5)
#             outputs = predictions(mean, mean_bias, np.exp(log_std), np.exp(log_std_bias),\
#                                   tau,nu,sigma_tau,sigma_nu,c_mu,c_std, inputs_test)
            rmse = acc_rmse(targets_test,outputs,train_std,train_mu)
            rmse_master = np.append(rmse_master,rmse)
            tau_master = np.append(tau_master, tau)
            tau_std_master = np.append(tau_std_master, sigma_tau)
            mean_master = np.append(mean_master,mean)
            nu_master = np.append(nu_master,nu)
            nu_std_master = np.append(nu_std_master, sigma_nu)
            lambda_master = np.append(lambda_master, lambda_b_q)
            theta_master = np.append(theta_master, theta_b_q)
            preci_master = np.append(preci_master, preci_mu)
            preci_log_std_master = np.append(preci_log_std_master, np.exp(preci_log_std))
            c_mu_master = np.append(c_mu_master,c_mu)
            c_std_master = np.append(c_std_master,c_std)
            print("Iteration: {} lower bound: {} test rmse: {}".format(t, -objective(params, t),rmse))
            plt.plot(theta_master[-50:])
            #plt.plot(preci_master[-20:])
            plt.show()
            plt.pause(1.0/60.0)
    if init == True:
        init_var_params = np.loadtxt('init_1e-5_shrink.txt')
    else:
        init_var_params,lambda_a_q, theta_a_q, lambda_b_q, theta_b_q = inialization(shapes)
        
    print("Optimizing variational parameters...")
    variational_params = adam(gradient, init_var_params,
                              step_size = step, num_iters = max_iter, callback=callback)
```
